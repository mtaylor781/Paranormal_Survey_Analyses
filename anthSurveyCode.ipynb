```python
print(len(df.index))
fdf = df[df['Finished'] == 'True']
fdf = fdf.dropna()
print(len(fdf.index))
```

```python
fdf = fdf.replace(['Never'], 0)
fdf = fdf.replace(['Less than once a month, but at least once a year'], 1)
fdf = fdf.replace(['Once or twice a month'], 2)
fdf = fdf.replace(['Once or twice a week'], 5)
fdf = fdf.replace(['Most days'], 8)
fdf = fdf.replace(['Every day'], 10)
fdf = fdf.astype({"Q4_1":"float","Q4_2":"float","Q4_3":"float","Q4_4":"float","Q4_5":"float","Q4_6":"float","Q4_7":"float","Q4_8":"float","Q4_9":"float","Q4_10":"float","Q4_11":"float"})

fdf = fdf.replace(['Less than 1 hour'], 1)
fdf = fdf.replace(['1-2 hours'], 2)
fdf = fdf.replace(['3-4 hours'], 4)
fdf = fdf.replace(['5-6 hours'], 6)
fdf = fdf.replace(['7-8 hours'], 8)
fdf = fdf.replace(['9+ hours'], 10)
fdf = fdf.astype({"Q5_3":"float","Q5_2":"float","Q5_3":"float"})

fdf = fdf.replace(['Strongly disagree'], 0)
fdf = fdf.replace(['Mildly disagree'], 1)
fdf = fdf.replace(['Mildly agree'], 3)
fdf = fdf.replace(['Strongly agree'], 5)
fdf = fdf.astype({"Q8_1":"float","Q8_2":"float","Q8_3":"float","Q8_4":"float","Q8_5":"float","Q8_6":"float","Q8_7":"float"})

fdf = fdf.replace(['Have never heard of this'], 0)
fdf = fdf.replace(['Extremely unlikely'], 0)
fdf = fdf.replace(['Somewhat unlikely'], 1)
fdf = fdf.replace(['Somewhat likely'], 3)
fdf = fdf.replace(['Extremely likely'], 5)
fdf = fdf.astype({"Q9_1":"float","Q9_2":"float","Q9_3":"float","Q9_4":"float","Q9_5":"float","Q9_6":"float","Q9_7":"float","Q9_8":"float","Q9_9":"float"})

fdf = fdf.replace(['Not Afraid'], 0)
fdf = fdf.replace(['Slightly Afraid'], 3)
fdf = fdf.replace(['Afraid'], 4)
fdf = fdf.replace(['Very Afraid'], 6)
fdf = fdf.astype({"Q10_1":"float","Q10_2":"float","Q10_3":"float","Q10_4":"float","Q10_5":"float","Q10_6":"float","Q10_7":"float","Q10_8":"float"})

fdf = fdf.astype({"Q11_1":"float","Q11_2":"float","Q11_3":"float","Q11_4":"float","Q11_5":"float","Q11_6":"float","Q11_7":"float","Q11_8":"float","Q11_9":"float","Q11_10":"float","Q11_11":"float","Q11_12":"float"})

fdf = fdf.replace(['Strongly Disagree'], 0)
fdf = fdf.replace(['Disagree'], 1)
fdf = fdf.replace(['Agree'], 4)
fdf = fdf.replace(['Strongly Agree'], 6)
fdf = fdf.astype({"Q12_1":"float","Q12_2":"float","Q12_3":"float","Q12_4":"float","Q12_5":"float"})

fdf = fdf.astype({"Q13_1":"float","Q13_2":"float","Q13_3":"float","Q13_4":"float","Q13_5":"float","Q13_6":"float","Q13_7":"float","Q13_8":"float","Q13_9":"float","Q13_10":"float"})

fdf = fdf.astype({"Q14_1":"float","Q14_2":"float","Q14_3":"float","Q14_4":"float","Q14_5":"float","Q14_6":"float","Q14_7":"float"})

fdf = fdf.replace(['Not at all religious'], 0)
fdf = fdf.replace(['Not too religious'], 2)
fdf = fdf.replace(['Somewhat religious'], 3)
fdf = fdf.replace(['Very religious'], 6)
fdf = fdf.astype({'Q15':'float'})

fdf = fdf.replace(['Only on special occasions (e.g., weddings, funerals, holidays)'], 1)
fdf = fdf.replace(['Less than once a year'], 0)
fdf = fdf.replace(['Once or twice a year'], 1)
fdf = fdf.replace(['Several times a year'], 2)
fdf = fdf.replace(['Once a month'], 5)
fdf = fdf.replace(['2-3 times a month'], 7)
fdf = fdf.replace(['Weekly'], 10)
fdf = fdf.replace(['Several times a week'], 13)
fdf = fdf.astype({'Q16':'float'})

fdf = fdf.replace(['I don\'t know.'], 0)
fdf = fdf.replace(['The Bible is an ancient book of history and legends.'], 0)
fdf = fdf.replace(['The Bible contains some human error.'], 3)
fdf = fdf.replace(['The Bible is perfectly true, but it should not be taken literally, word-for-word. We must interpret its meaning.'], 7)
fdf = fdf.replace(['The Bible means exactly what it says. It should be taken literally, word-for-word, on all subjects.'], 10)
fdf = fdf.astype({'Q17':'float'})

fdf = fdf.replace(['Under $20,000'], 2)
fdf = fdf.replace(['$20,000 to under $30,000'], 3)
fdf = fdf.replace(['$40,000 to under $50,000'], 5)
fdf = fdf.replace(['$50,000 to under $60,000'], 6)
fdf = fdf.replace(['$60,000 to under $70,000'], 7)
fdf = fdf.replace(['$70,000 to under $100,000'], 10)
fdf = fdf.replace(['$100,000 to under $150,000'], 15)
fdf = fdf.replace(['$150,000 or more'], 20)
fdf = fdf.replace(['Prefer not to answer'], 10)
fdf = fdf.astype({'Q28':'float'})

print(fdf.dtypes)
```

```python
sns.histplot(data=fdf, x="Q17")
```

```python
fdf['PsudoArchScore'] = fdf.loc[:,['Q8_1', 'Q8_2', 'Q8_3', 'Q8_4', 'Q8_5', 'Q8_6', 'Q8_7']].sum(axis=1)
fdf['PsudoArchScore'] = ( fdf['PsudoArchScore'] - fdf['PsudoArchScore'].mean() ) / fdf['PsudoArchScore'].std()

fdf['PsudoHealthScore'] = fdf.loc[:,['Q9_1', 'Q9_2', 'Q9_3', 'Q9_4', 'Q9_5', 'Q9_6', 'Q9_7', 'Q9_8', 'Q9_9']].sum(axis=1)
fdf['PsudoHealthScore'] = ( fdf['PsudoHealthScore'] - fdf['PsudoHealthScore'].mean() ) / fdf['PsudoHealthScore'].std()

fdf['internetTime'] = fdf.loc[:,['Q5_1', 'Q5_2', 'Q5_3']].sum(axis=1)
fdf['internetTime'] = ( fdf['internetTime'] - fdf['internetTime'].mean() ) / fdf['internetTime'].std()

fdf['discriminateScore'] = fdf.loc[:,['Q11_8', 'Q11_9', 'Q12_1', 'Q12_2', 'Q12_4']].sum(axis=1)
fdf['discriminateScore'] = ( fdf['discriminateScore'] - fdf['discriminateScore'].mean() ) / fdf['discriminateScore'].std()

fdf['governmentDistrust'] = fdf.loc[:,['Q13_1', 'Q13_2', 'Q13_3', 'Q13_4', 'Q13_5', 'Q13_6', 'Q13_7', 'Q13_8', 'Q13_9']].sum(axis=1)
fdf['governmentDistrust'] = ( fdf['governmentDistrust'] - fdf['governmentDistrust'].mean() ) / fdf['governmentDistrust'].std()

fdf['religous'] = fdf.loc[:,['Q15', 'Q16', 'Q17']].sum(axis=1)
fdf['religous'] = ( fdf['religous'] - fdf['religous'].mean() ) / fdf['religous'].std()

fdf['tvNews'] = fdf.loc[:,['Q4_3', 'Q4_4', 'Q4_5', 'Q4_6', 'Q4_7', 'Q4_11']].sum(axis=1)
fdf['tvNews'] = ( fdf['tvNews'] - fdf['tvNews'].mean() ) / fdf['tvNews'].std()

fdf['onlineNews'] = fdf.loc[:,['Q4_8', 'Q4_9']].sum(axis=1)
fdf['onlineNews'] = ( fdf['onlineNews'] - fdf['onlineNews'].mean() ) / fdf['onlineNews'].std()

fdf = fdf.rename(columns={"Q4_10": "socialNews", "Q10_6": "disinformationFear", "Q21": "majorDept", "Q22": "major", "Q24": "gender", "Q28": "ecBackground", "Q2": "Politics", "Q3": "Politics2"})

fdf['disinformationFear'] = ( fdf['disinformationFear'] - fdf['disinformationFear'].mean() ) / fdf['disinformationFear'].std()
fdf['socialNews'] = ( fdf['socialNews'] - fdf['socialNews'].mean() ) / fdf['socialNews'].std()
fdf['ecBackground'] = ( fdf['ecBackground'] - fdf['ecBackground'].mean() ) / fdf['ecBackground'].std()


agDf = fdf[['majorDept', 'major', 'gender', 'ecBackground', 'religous', 'Politics', 'Politics2', 'socialNews', 'tvNews', 'onlineNews', 'internetTime', 'governmentDistrust', 'disinformationFear', 'discriminateScore', 'PsudoArchScore', 'PsudoHealthScore']]
```

```python
heatMapDf = agDf
heatMapDf.Politics=heatMapDf.Politics.astype('category').cat.codes
heatMapDf.Politics2=heatMapDf.Politics2.astype('category').cat.codes


cormat = heatMapDf.corr()
round(cormat,2)
sns.heatmap(cormat, annot=True);
```
